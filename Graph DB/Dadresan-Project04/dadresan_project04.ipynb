```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('project04.csv', index_col=0)
data.shape
```

```python
# Create a medium-size graph as an adjacency matrix and save it as a .csv file. The graph
# must have at least 20-30 nodes, have weights, and have nodes of different in/out degrees
G = nx.DiGraph(data.values)
labels = nx.get_edge_attributes(G,'weight')
pos = nx.spring_layout(G)
nx.draw(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Shape 1 - Node Color green 
nx.draw(G, pos, with_labels=1,node_color='g')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Shape 2 - Layout Circular
pos=nx.circular_layout(G)
nx.draw(G,pos, with_labels=1, node_color='purple', )
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Shape 3 - Line Styles to dashed and the spiral layout
pos = nx.spiral_layout(G)
nx.draw(G, pos, with_labels=1,node_size = 900, node_color='b', alpha=0.5, edge_color='black', style = 'dashed', width = 2)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Shape 4 - Label Font and edge color
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=1,node_size = 900, node_color='r',width=1, alpha=0.5, edge_color='g')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Shape 5 - Shape
pos = nx.spring_layout(G)
nx.draw(G,pos, with_labels=1,font_size=20,node_color='cyan', node_size = 900, edge_color = 'magenta')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```
